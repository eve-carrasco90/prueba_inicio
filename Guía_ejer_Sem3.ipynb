```python
import pandas as pd
```

```python
df = pd.read_csv("rrhh_2.csv")
df
```

```python
def calculo_precio_producto(Experiencia, Hijos):
    return Experiencia * Hijos
```

```python
import matplotlib.pyplot as plt
plt.bar(df["Educacion"], df["Hijos"], color="skyblue", edgecolor="black")
plt.title("Gráfico de Barras")
plt.xlabel("Educacion")
plt.ylabel("Hijos")
plt.show()
```

```python
df["Precio_Producto"] = df.apply(lambda x: calculo_precio_producto(x["Experiencia"], x["Hijos"]), axis=1)
df
```
